```python

from scipy.stats import boxcox

def reduce_skewness(column):
    # Calculate the skewness of the column
    skew_value = column.skew()

    # Check if the skewness is above a certain threshold
    if abs(skew_value) > 0.5:
        # Apply the appropriate transformation based on skewness direction
        if skew_value > 0:
            # Apply log transformation
            column = np.log(column)
        else:
            # Apply Box-Cox transformation
            column, _ = boxcox(column)

    return column

# Apply the reduce_skewness function to selected columns
columns_to_transform = ['nox', 'rm', 'age', 'dis', 'rad', 'tax', 'medv']
df[columns_to_transform] = df[columns_to_transform].apply(reduce_skewness)
```

```python
sns.pairplot(df)
```

```python
df=df.drop(['nox','rm'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(['medv'],axis=1)
y=df['medv']

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)

print("coefficient is",lr.coef_)
print("intercept is",lr.intercept_)
print("fesures in linear",lr.n_features_in_)
print("features :",lr.feature_names_in_)
```

```python
lr.score(X, y, sample_weight=None)

```
